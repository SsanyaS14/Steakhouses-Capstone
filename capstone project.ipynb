Get Data From Wikipedia using Soap method and using pandas to make Data Frame

```python
import pandas as pd

import requests

from bs4 import BeautifulSoup

req = requests.get("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")

soup = BeautifulSoup(req.content,'lxml')

table = soup.find_all('table')[0]

df = pd.read_html(str(table))

data=pd.DataFrame(df[0])
```

```python
data.head()
```

Use Sorting to rid of Not Assigned values

```python
data=data[data.Borough != 'Not assigned']
data = data.sort_values(by=['Postcode','Borough'])

data.reset_index(inplace=True)
data.drop('index',axis=1,inplace=True)

data.head()
```

```python
grouped = data.groupby(['Postcode', 'Borough'], as_index=False, sort=False).agg(','.join)
```

```python
grouped.head()
```

```python
data_2 = pd.read_csv("http://cocl.us/Geospatial_data")
```

```python
Latitude = data_2['Latitude']
Longitude = data_2['Longitude']
```

```python
grouped['Latitude'] = Latitude
grouped['Longitude'] = Longitude
```
